# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection & Pre-Processing

```python
raw_mail_data = pd.read_csv('sms_data.csv')
```

```python
print(raw_mail_data)
```

```python
raw_mail_data.head()
```

```python
raw_mail_data.isna().sum() 
```

```python
raw_mail_data.shape
```

# Label Encoding

```python
# label spam mail as 0;  ham mail as 1;

raw_mail_data.loc[raw_mail_data['Category'] == 'spam', 'Category',] = 0
raw_mail_data.loc[raw_mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separating the data as texts and label

X = raw_mail_data['Message']

Y = raw_mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

# Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase='True')

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

# Training the Model

# Logistic Regression  - Algo 1

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data: ',accuracy_on_test_data)
```

# KNN - Algorithm 2

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train_features, Y_train)
print("Accuracy on train data : ",model.score(X_train_features, Y_train))
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data: ',accuracy_on_test_data)
```

# Decision Tree - Algorithm 3

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train_features, Y_train)
print("Accuracy on train data : ",model.score(X_train_features, Y_train))
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data: ',accuracy_on_test_data)
```

# Comparison between the three Algorithm
